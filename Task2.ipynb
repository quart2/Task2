```python
import yfinance as yf
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np

# Step 1: Load stock data (e.g., Apple)
stock = yf.download('AAPL', start='2023-01-01', end='2024-12-31')  # You can change to any stock
stock = stock[['Close']]  # Use only 'Close' prices

# Step 2: Create label column (closing price of next day)
stock['Target'] = stock['Close'].shift(-1)

# Step 3: Drop last row (has NaN in Target)
stock = stock[:-1]

# Step 4: Features and labels
X = stock[['Close']]
y = stock['Target']

# Step 5: Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# Step 6: Train model
model = LinearRegression()
model.fit(X_train, y_train)

# Step 7: Predict
predictions = model.predict(X_test)

# Step 8: Plot actual vs predicted
plt.plot(y_test.values, label="Actual")
plt.plot(predictions, label="Predicted")
plt.title("Stock Price Prediction (Next Day)")
plt.xlabel("Days")
plt.ylabel("Price ($)")
plt.legend()
plt.show()
```

```python
import yfinance as yf
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np

# Step 1: Load Tesla stock data
stock = yf.download('TSLA', start='2024-01-01', end='2025-1-31')
stock = stock[['Close']]  # Use only 'Close' prices

# Step 2: Predict next day's closing price
stock['Target'] = stock['Close'].shift(-1)

# Step 3: Drop last row (NaN in Target)
stock = stock[:-1]

# Step 4: Features and labels
X = stock[['Close']]
y = stock['Target']

# Step 5: Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# Step 6: Train Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Step 7: Predict
predictions = model.predict(X_test)

# Step 8: Plot
plt.plot(y_test.values, label="Actual")
plt.plot(predictions, label="Predicted")
plt.title("Tesla Stock Price Prediction (Next Day)")
plt.xlabel("Days")
plt.ylabel("Price ($)")
plt.legend()
plt.show()
```
